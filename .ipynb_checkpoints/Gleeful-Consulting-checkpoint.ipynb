```python
import pandas as pd

loc = "GleeClubData.xlsx"

df = pd.read_excel(loc)
```

```python
df
```

```python
# Change the column names into something more manageable.

df.columns = ['size', 'teamsexratio', 'enrolled', 'schoolsexratio', 'popserved', 'numpract', 'numcompete', 'score']
```

```python
import statsmodels.formula.api as sm

df.corr()
```

```python
import numpy as np

# create a dummy variable that takes the value of 1 when score is greater than or equal to 4.25
df["winner"] = np.where((df["score"]>= 4.25), 1, 0)
```

### Variables that are highly correlated with score are: enrolled, popserved, numpract, and numcompete.

```python
#check for na, spoiler alert there are none

df.dropna().head()
```

```python
#just for kicks regress everything anyway
reg1 = sm.ols(formula='score ~ size + teamsexratio + enrolled + schoolsexratio + popserved + numpract + numcompete', data=df).fit()
reg1.summary()
```

#### Regress again without an intercept, since truth be told it doesn't make much sense for there to be any score if any of the independent variables were zero. Who has ever heard of a glee team from a school with zero enrollment getting a score?

#### And this time we are taking out variables that don't correlate with score and only using: enrolled, popserved, numpract, and numcompete

```python
reg2 = sm.ols(formula='score ~ enrolled + popserved + numpract + numcompete - 1', data=df).fit()
reg2.summary()
```

#### According to this analysis, each person from the population served by the team's school increases the linear probability of having a score greater than or equal to 4.25 by a value that is signficant, but suuuuuper small to the point where we have to use scientific notation to write it. We could just not include this in our regression, but that would be a waste of a significant predictor of score being equal to or above 4.25. Instead lets scale the population into thousands and see what the impact of having 1,000 more people in a school district is on score being equal to or above 4.25.

```python
df["popservedperthou"] = df["popserved"]/1000

reg3 = sm.ols(formula='score ~ enrolled + popservedperthou + numpract + numcompete - 1', data=df).fit()
reg3.summary()
```

### Results
#### Below are the full coefficients of our model, using them we can create a function that predicts the linear probability of getting a score of greater than or equal to 4.25 based on the number of people enrolled in the school, the population served by the school in thousands, the number of practices the glee team has had, and the number of competitions the glee team has participated in.

```python
reg3.params
```

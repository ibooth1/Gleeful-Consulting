```python
import pandas as pd

loc = "fallcampdata/EarthCrisisData.xlsx"
df = pd.read_excel(loc)

headers = list(df.columns.values)
headers[0] = 'year'
headers[1] = 'waste'
df.columns = headers
df.head()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='waste ~ year',data=df).fit()
result.summary()
```

```python
import pandas as pd

loc="fallcampdata/TamiamiTamalesData.xlsx"

df = pd.read_excel(loc)
df = df.dropna()
df.head()
```

```python
df.columns = ["loc", "sales", "emp", "rest", "othcrt", "price"]
df.head()
```

```python
df.corr()
```

```python
import statsmodels.formula.api as sm
#result = sm.ols(formula='waste ~ year',data=df).fit()
mod = sm.ols(formula = ' sales ~ emp + rest + othcrt + price', data=df).fit()
mod.summary()
```
